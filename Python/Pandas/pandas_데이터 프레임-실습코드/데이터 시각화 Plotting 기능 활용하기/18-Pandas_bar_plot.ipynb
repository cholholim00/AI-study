## bar plot

* 막대 그래프(-graph), 바 차트(bar chart), 바 그래프(bar graph)는 표현 값에 비례하여 높이와 길이를 지닌 직사각형 막대로 범주형 데이터를 표현하는 차트나 그래프를 말한다. 막대는 수직으로나 수평으로 그릴 수 있다. 수직 막대 그래프는 선 그래프 또는 라인 그래프(line graph)라고도 한다. (출처 : [막대 그래프 - 위키백과, 우리 모두의 백과사전](https://ko.wikipedia.org/wiki/%EB%A7%89%EB%8C%80_%EA%B7%B8%EB%9E%98%ED%94%84))
* https://pandas.pydata.org/pandas-docs/stable/user_guide/visualization.html

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline

# 한글폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 1. 실습 데이터 준비
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
ts.head()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=ts.index, columns=list('ABCD'))
df.head(6)
```

```python
# 2. tail(2): 뒤에서부터 2개의 행 확인
df.tail(2)
```

```python
# 3. iloc[5]: 인덱스 순서상 5번째 행(Row) 하나만 가져오기
df.iloc[5]
```

```python
# 4. plot(kind='bar'): 가져온 1개 행의 데이터로 수직 막대 그래프 그리기
df.iloc[5].plot(kind='bar')
```

```python
# 5. plot.bar(): 위와 동일한 기능 (메서드 방식)
df.iloc[5].plot.bar()
# 6. plt.axhline(0, color='k'):
# y=0 위치에 검은색(k) 수평선(axhline)을 그어 기준선을 만듭니다.
plt.axhline(0, color='k')
```

```python
# --- 여러 데이터 막대 그래프 ---
```

```python
# 7. 새 데이터프레임 생성 (10행 4열)
df2 = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df2.head(3)
```

```python
# 8. df2.plot.bar(): 10개 행, 4개 열 데이터를 모두 막대 그래프로 그리기
df2.plot.bar()
```

```python
# 9. stacked=True: 막대를 옆으로 펼치지 않고 위로 쌓아 올리기 (누적 막대)
df2.plot.bar(stacked=True)
```

```python
# 10. barh(): 수평(Horizontal) 막대 그래프 그리기 (누적 옵션 포함)
df2.plot.barh(stacked=True)
```
