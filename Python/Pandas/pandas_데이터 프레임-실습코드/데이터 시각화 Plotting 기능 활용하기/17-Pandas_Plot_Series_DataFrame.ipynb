```python
# --- Series 시각화 ---
```

```python
# 1. pd.Series(): 1차원 데이터 생성
# np.random.randn(1000): 표준정규분포 난수 1000개 생성
# index=pd.date_range(...): 2000년 1월 1일부터 1000일간의 날짜를 인덱스로 지정
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('1/1/2000', periods=1000))
```

```python
# 2. ts.plot(): 기본 선 그래프 그리기 (x축: 날짜, y축: 값)
ts.plot()
```

```python
# 3. cumsum(): 누적 합계 구하기 (Cumulative Sum)
# 난수를 계속 더해서 주식 차트처럼 연속적인 흐름을 만듭니다.
ts = ts.cumsum()
```

```python
# 4. 누적된 데이터로 다시 그래프 그리기
ts.plot()
```

```python
# --- DataFrame 시각화 ---
```

```python
# 5. pd.DataFrame(): 2차원 데이터 생성 (1000행 4열)
# columns=list('ABCD'): 컬럼 이름을 A, B, C, D로 지정
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=ts.index, columns=list('ABCD'))
```

```python
# 6. df.plot(): 데이터프레임의 모든 컬럼(A,B,C,D)을 한 번에 선 그래프로 그리기
df.plot()
```

```python
# 7. 데이터프레임도 누적 합계(cumsum)를 구해서 흐름 만들기
df = df.cumsum()
```

```python
# 8. 누적된 데이터프레임 그리기 (4개의 선이 그려짐)
df.plot()
```

```python
# --- X, Y축 지정해서 그리기 ---
```

```python
# 9. 새로운 데이터프레임 생성 및 누적
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
```

```python
# 10. head(): 상위 5개 데이터 확인
df3.head()
```

```python
# 11. 새로운 컬럼 'A' 추가
# 0부터 999까지의 숫자를 Series로 만들어 'A' 컬럼에 넣습니다.
df3['A'] = pd.Series(list(range(len(df))))
```

```python
# 11-1. 5개 데이터 확인
df3.head()
```

```python
# 12. df3.plot(x='A', y='B'):
# x축을 'A' 컬럼값으로, y축을 'B' 컬럼값으로 지정해서 그립니다.
df3.plot(x='A', y='B')
```
