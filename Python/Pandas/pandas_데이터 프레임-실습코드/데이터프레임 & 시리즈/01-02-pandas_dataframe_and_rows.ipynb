```python
import pandas as pd
```

```python
df = pd.DataFrame( # 데이터 프레임 생성
        {"a" : [4 ,5, 6],
        "b" : [7, 8, 9],
        "c" : [10, 11, 12]},
        index = [1, 2, 3])
```

```python
df # 전체 컬럼 조회
# df["a"] # 특정 컬럼 조회하기
# df[["a", "b"]] # 두가지 이상 컬럼 조회하기
```

```python
df.loc[3, "a"] # 3번째 low 데이터에 조회
# df.loc[3, "b"] # 3번째 b에 해당하는 low 데이터에 조회 (행,열 순으로)
```

```python
df.loc[[1, 2], ["a", "b"]] # 2,3번째 a,b에 해당하는 low 데이터에 조회 (행,열 순으로)
```

```python
df = pd.DataFrame( # 시리즈 데이터 생성
        [[4, 7, 10],
        [5, 8, 11],
        [6, 9, 12]],
        index=[1, 2, 3],
        columns=['a', 'b', 'c']) # 컬럼 이름으로 따로 지정
df
```

```python
import numpy as np
df = pd.DataFrame(
        {"a" : [4 ,5, 6, 6, np.nan],
        "b" : [7, 8, np.nan, 9, 9],
        "c" : [10, 11, 12, np.nan, 12]},
        index = pd.MultiIndex.from_tuples(
        [('d',1),('d',2),('e',2),('e',3),('e',4)], #
        names=['n','v'])) # 2개의 인덱스를 가지는 데이터 프레임
df
```
