1. 실습 데이터 준비
먼저 Seaborn 라이브러리에 내장된 자동차 연비 데이터(mpg)를 불러옵니다.

```python
import pandas as pd
import seaborn as sns
```

```python
# 1. 자동차 연비 데이터셋 로드
df = sns.load_dataset("mpg")
# 2. 상위 3개 행만 미리보기
df.head(3)
```

2. GroupBy (그룹화하여 요약하기)
데이터를 특정 기준(예: 제조국가)으로 묶은 뒤, 합계나 평균 등 통계량을 계산합니다.

```python
# 3. 그룹별 개수 세기 (Size)
# df.groupby(by="그룹기준컬럼")["계산할컬럼"].함수()
# 'origin'(제조국)별로 그룹을 짓고, 'cylinders' 컬럼의 데이터 개수(size)를 셉니다.
# 결과: europe 70개, japan 79개, usa 249개
df.groupby(by="origin")['cylinders'].size()
```

```python
# 4. value_counts와 비교
# 위에서 구한 size()는 사실 value_counts()와 결과가 같습니다.
# 둘 다 각 범주의 빈도수를 세는 역할을 합니다.
df["origin"].value_counts()
```

```python
# 5. 여러 기준으로 묶어서 평균 구하기 (Multi-Index)
# 'model_year'(연식)와 'origin'(제조국) 두 가지 기준으로 묶습니다.
# 그 그룹들 각각에 대해 'cylinders'(실린더 수)의 평균(mean)을 구합니다.
# pd.DataFrame()으로 감싸주면 보기 좋은 표 형태로 출력됩니다.
pd.DataFrame(df.groupby(['model_year', 'origin'])['cylinders'].mean())
```

3. Shift와 Rank (이동과 순위)
데이터의 순서나 위치를 기반으로 연산하는 기능들입니다. 실습을 위해 간단한 df2를 만듭니다.

```python
# 6. 실습용 데이터프레임 df2 생성
df2 = pd.DataFrame(
[[4, 7, 10],
[5, 11, 8],
[6, 9, 12]],
index=[1, 2, 3],
columns=['a', 'b', 'c'])
```

```python
# 데이터 확인
df2
```

```python
# 7. 데이터 밀어내기 (Shift)
# 'b' 컬럼의 데이터를 아래로 2칸씩 밀어냅니다.
# 밀려나서 빈 위쪽 2칸은 NaN(결측치)으로 채워집니다.
# 주로 '어제 가격', '지난달 매출' 등 과거 데이터와 비교할 때 씁니다.
df2['b'].shift(2)
```

```python
# 8. 순위 구하기 (Rank)
# 'model_year' 컬럼의 값들에 대해 순위(등수)를 매깁니다.
# method='first': 값이 같을 경우(동점), 먼저 나온 행에 더 높은 등수(작은 숫자)를 줍니다.
# 예: 70년식이 여러 개면 먼저 등장한 게 1등, 그 다음이 2등...
df["model_year"].rank(method='first').head()
```

```python
# 데이터 확인
df2
```

4. 누적 함수 (Cumulative Functions)
처음부터 현재 행까지 쭉 쌓아가면서 계산하는 함수들입니다. (예: 누적 매출, 누적 방문자 수 등)

```python
# 9. 누적 합계 (Cumulative Sum)
# 첫 행부터 아래로 내려가면서 값을 계속 더합니다.
# 예: a컬럼 -> 4, 4+4=8, 8+5=13...
df2.cumsum()
```

```python
# 10. 누적 최댓값 (Cumulative Max)
# 내려가면서 지금까지 본 값 중 가장 큰 값을 기록합니다.
# 예: a컬럼 -> 4, 4, 5(갱신), 5, 5
df2.cummax()
```

```python
# 11. 누적 최솟값 (Cumulative Min)
# 내려가면서 지금까지 본 값 중 가장 작은 값을 기록합니다.
# 예: a컬럼 -> 4, 4, 4, 4, 1(갱신)
df2.cummin()
```

```python
# 12. 누적 곱 (Cumulative Product)
# 내려가면서 값을 계속 곱합니다.
# 예: a컬럼 -> 4, 4*4=16, 16*5=80...
df2.cumprod()
```
